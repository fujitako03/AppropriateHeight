## dreamlike-photoreal-2.0 とは？
Dreamlike Photoreal 2.0は、Stable Diffusion 1.5をベースにしたリアルな写真風画像を生成するモデル

## 利用上の注意
- ご注意このモデルはムラムラしています。NSFWを回避したい場合は、ネガティブプロンプトに`nude, naked`を追加してください。
- プロンプトに`photo`を追加することで、より写実的に見えるようにすることができます
- プロンプトによっては、非正方形のアスペクト比の方がうまくいく場合があります。縦長の写真が必要な場合は、縦長のアスペクト比を使用してみてください。横長の写真が欲しい場合は、横長のアスペクト比を使用してみてください。
- このモデルは768x768pxの画像で学習させたので、768x768px、640x896px、896x640pxなどを使ってみてください。また、768x1024pxや1024x768pxといった高解像度でも、かなり良好に動作します。

## とりあえず動かしてみる

```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "dreamlike-art/dreamlike-photoreal-2.0"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "photo, a church in the middle of a field of crops, bright cinematic lighting, gopro, fisheye lens"
image = pipe(
    prompt).images[0]

out_path = "./result.jpg"
image.save( out_path)

# 表示
from IPython.display import Image,display
display(Image(out_path))
```

## 人物写真を生成してみる

杖を持った、大きな帽子を被った、青い目の、年老いた男の魔法使い

```python
prompt = "photo, a old man wizard, big hat , in black robe, with wand, blue eyes"
image = pipe(
    prompt,
    num_inference_steps=25
    ).images[0]

out_path = "./result.jpg"
image.save( out_path)

# 表示
from IPython.display import Image,display
display(Image(out_path))
```
